# Prediction of Canada per capita income

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Loadind data from csv file
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
#Plotting scatter graph
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income_per_capita')
plt.scatter(df["year"],df["per capita income (US$)"],color='red',marker='+')
```

```python
#Creating linear regression object
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
reg.predict([[2020]])
```

```python
#Checking internal calculations
reg.coef_
reg.intercept_
```

```python
reg.predict([[2024]])
```

```python
#predicting file of years and storing the predicted values
d=pd.read_csv("years.csv")
p=reg.predict(d)
d['predicted per capita income']=p
d.to_csv("prediction.csv",index=False)
```

```python
#Plotting line of regression
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income_per_capita')
plt.scatter(df["year"],df["per capita income (US$)"],color='red',marker='+')
plt.plot(df['year'],reg.predict(df[['year']]),color="blue")
```
